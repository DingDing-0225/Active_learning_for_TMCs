```python
%run eval.ipynb
import pickle
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score, make_scorer
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
# 加载数据
data = pd.read_csv('data/compound_feat.csv').iloc[:,1:]

# scaler = MinMaxScaler()
# data = scaler.fit_transform(data.values)
# data_x, data_y = data[:,:-1], data[:, -1:]

print(data.shape)

data_x, data_y = data.iloc[:, :-1].values, data.iloc[:, -1:].values

# 拆分训练测试集
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=21)
```

```python
# 自定义评估函数
def error_ratio(y_test, y_pred):
    y_test = np.array(y_test).flatten()
    y_pred = np.array(y_pred)
    ratio = []
    for t, p in zip(y_test, y_pred):
        r = abs(p - t)/t
        ratio.append(r)
    return str(round(np.mean(ratio)*100, 2))+'%'
```

```python
from sklearn.linear_model import ElasticNet
import sklearn
elastic_model = ElasticNet(alpha=0.00001, l1_ratio=0.001, random_state=42)

# 数据标准化
scaler = sklearn.preprocessing.StandardScaler()
scaler.fit(X_train)
X_std = scaler.transform(X_train)

# 训练模型
elastic_model.fit(X_std, y_train)

# elastic_y_pred = elastic_model.predict(X_test)
# elastic_r2score = r2_score(y_test, elastic_y_pred)
# elastic_mse = mean_squared_error(y_test, elastic_y_pred)
# elastic_rmse = np.sqrt(elastic_mse)
# elastic_error_ratio = error_ratio(y_test, elastic_y_pred)
# print(f'elastic_model: \n r2: {elastic_r2score} \n rmse: {elastic_rmse} \n mse:{elastic_mse} \n error_ratio:{elastic_error_ratio}')

# # 交叉验证
# mse_scorer = make_scorer(mean_squared_error)
# score = get_cross_val_score(elastic_model, data_x, data_y, scoring=mse_scorer)
# print('交叉验证mse：',[round(abs(s),4) for s in score])
# print('交叉验证rmse：',[round(np.sqrt(abs(s)),4) for s in score])

# # 保存模型
# with open('model/elastic_net.pkl', 'wb') as f:
#     pickle.dump(elastic_model, f)
# print('save model successful')
```

```python
import shap
# 对模型文件elastic_model进行解释
explainer = shap.explainers.Linear(elastic_model, X_std)
# 传入特征矩阵X，计算shap值
shap_values = explainer(X_std)
print(shap_values.shape)
for i in range(5):
    shap.plots.scatter(shap_values[:, i])
    break
```

```python
# 转换原始特征值
cols = np.array(['AB_AN', 'AB_AW', 'AB_CR', 'AB_AR', 'AB_VR', 'AB_AV', 'AB_FIE',
       'AB_SIE', 'AB_EA', 'AB_EN', 'AB_Nd', 'AB_Nv', 'AB_D', 'AB_EC',
       'AB_TC', 'C_AN', 'C_AW', 'C_CR', 'C_AR', 'C_VR', 'C_AV', 'C_FIE',
       'C_SIE', 'C_EA', 'C_EN', 'C_Np', 'C_Nv', 'C_D', 'CurrentDensity'])
for i, c in enumerate(cols):
    shap_values.feature_names[i] = c
    
shap_values.data = X_train

x_df = pd.DataFrame(X_train, columns=cols)

cols_ = cols

# shap.plots.scatter(shap_values[:, 0])
shap.summary_plot(shap_values, x_df[cols_])
```
