```python
# 加载数据
# 分层取样
data = pd.read_csv('data/compound_feat3.csv').iloc[:,1:]
bins = [-np.inf,0.6,0.7,0.8,0.9,1,1.1,1.2,1.4,1.6,1.8,2,np.inf]
data['bin'] = pd.cut(data['label'], bins=bins, labels=False)
sample_size_per_bin = 25
def sample_with_minimum(group, sample_size, random_state=0):
    return group.sample(min(len(group), sample_size), random_state=random_state)
train = data.groupby('bin').apply(lambda x: sample_with_minimum(x, sample_size_per_bin)).reset_index(drop=True)

train_indices = train.index
test = data.drop(train_indices).reset_index(drop=True)
train = train.drop(columns='bin')
test = test.drop(columns='bin')

X_train, y_train = train.iloc[:,:-1].values, train.iloc[:,-1:].values
X_test, y_test = test.iloc[:,:-1].values, test.iloc[:,-1:].values
print(train.shape)
print(test.shape)
```

```python
# 过采样+分层采样（每一层按照比例抽取）
# 先进行过采样再进行分层采样
np.random.seed(42)
data = pd.read_csv('data/compound_feat4.csv').iloc[:,1:]
samples = data[data['label']<1]
new_data = pd.concat([data, samples], axis=0)
new_data = pd.concat([new_data, samples], axis=0)
new_data = new_data.sample(frac=1).reset_index(drop=True)

new_data['target'] = pd.qcut(new_data['label'], q=10, labels=False)
train_df, test_df = train_test_split(new_data, test_size=0.2, stratify=new_data['target'], random_state=42)

# 删除分箱列
train_df = train_df.drop(columns=['target'])
test_df = test_df.drop(columns=['target'])
X_train, y_train = train_df.iloc[:, :-1], train_df.iloc[:, -1:].values
X_test, y_test = test_df.iloc[:, :-1], test_df.iloc[:, -1:].values
print('train data:', X_train.shape)
print('X_test data:', X_test.shape)
```

```python
# 分层采样 + 过采样
# 先进行分层采样再进行过采样
np.random.seed(42)
data = pd.read_csv('data/compound_feat62.csv').iloc[:,1:]
data['target'] = pd.qcut(data['label'], q=10, labels=False)
train_df, test_df = train_test_split(data, test_size=0.2, stratify=data['target'], random_state=42)

#对训练集中的标签小于1的数据进行过采样
samples = train_df[train_df['label'] < 1.2]
train_df = pd.concat([train_df, samples], axis=0)
#train_df = pd.concat([train_df, samples], axis=0)
train_df = train_df.sample(frac=1).reset_index(drop=True)

# 删除分箱列
train_df = train_df.drop(columns=['target'])
test_df = test_df.drop(columns=['target'])

# 准备训练和测试数据
X_train, y_train = train_df.iloc[:, :-1], train_df.iloc[:, -1:].values
X_test, y_test = test_df.iloc[:, :-1], test_df.iloc[:, -1:].values

# 打印数据集形状
print('train data:', X_train.shape)
print('X_test data:', X_test.shape)
```

```python
# 过采样+随机划分数据集
data = pd.read_csv('data/compound_feat3.csv').iloc[:,1:]
samples = data[data['label']<1]
new_data = pd.concat([data, samples], axis=0)
new_data = pd.concat([new_data, samples], axis=0)
new_data = new_data.sample(frac=1).reset_index(drop=True)

data_x, data_y = new_data.iloc[:, :-1].values, new_data.iloc[:, -1:].values

X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=42)
print('train data:', X_train.shape)
print('X_test data:', X_test.shape)
```

# 1.决策树

```python
# 决策树
DecisionTree_model = DecisionTreeRegressor(random_state=25)
DecisionTree_model.fit(X_train, y_train)
DecisionTree_train_y_pred = DecisionTree_model.predict(X_train)
DecisionTree_y_pred = DecisionTree_model.predict(X_test)

# train
DecisionTree_r2score, DecisionTree_mse, DecisionTree_rmse, DecisionTree_error_ratio = eval(y_train, DecisionTree_train_y_pred)
print(f'DecisionTree: \n r2: {DecisionTree_r2score}  rmse: {DecisionTree_rmse}  mse: {DecisionTree_mse}  error_ratio:{DecisionTree_error_ratio}')

# test
DecisionTree_r2score, DecisionTree_mse, DecisionTree_rmse, DecisionTree_error_ratio = eval(y_test, DecisionTree_y_pred)
print(f'DecisionTree: \n r2: {DecisionTree_r2score}  rmse: {DecisionTree_rmse}  mse: {DecisionTree_mse}  error_ratio:{DecisionTree_error_ratio}')

# 交叉验证
#mse_scorer = make_scorer(mean_squared_error)
#score = get_cross_val_score(DecisionTree_model, data_x, data_y, scoring=mse_scorer)
#print('交叉验证mse：',[round(abs(s),4) for s in score])
#print('交叉验证rmse：',[round(np.sqrt(abs(s)),4) for s in score])

# 可视化
visual(DecisionTree_model, 'dt', X_train, X_test, y_train, y_test)

# 保存图像数据
save('desicionTree', '61', DecisionTree_train_y_pred, DecisionTree_y_pred, y_train, y_test)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

# 指定参数范围，最大深度，特征选择标准，子节点往下分裂所需的最小样本数
DT_param = {'max_depth':[3,5,7,9,10,11],
              'max_leaf_nodes':[25,30,40,50,60,70],
              'min_samples_leaf':[2,3,4,5,6,7,8],
}

# 构建决策树模型
DT_model = DecisionTreeRegressor(random_state=25)
# 网格搜索
DT_grid_search = GridSearchCV(DT_model,DT_param,scoring='neg_mean_squared_error',cv=10)
DT_grid_search.fit(X_train,y_train)

# 输出参数的最优值
print('最优参数组合：', DT_grid_search.best_params_)
print('最优参数得分mse：', abs(DT_grid_search.best_score_))

DT_train_y_pred = DT_grid_search.predict(X_train)
DT_y_pred = DT_grid_search.predict(X_test)

# trian
DT_r2score, DT_mse, DT_rmse, DT_error_ratio = eval(y_train, DT_train_y_pred)
print(f'DecisionTree: \n r2: {DT_r2score}  rmse: {DT_rmse}  mse: {DT_mse}  error_ratio: {DT_error_ratio}')

# test
DT_r2score, DT_mse, DT_rmse, DT_error_ratio = eval(y_test, DT_y_pred)
print(f'DecisionTree: \n r2: {DT_r2score}  rmse: {DT_rmse}  mse: {DT_mse}  error_ratio: {DT_error_ratio}')

# 可视化
visual(DT_grid_search, 'tuning_dt', X_train, X_test, y_train, y_test)

save('desicionTree_tuning', '61', DT_train_y_pred, DT_y_pred, y_train, y_test)
```

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import time
from IPython.display import display, clear_output
torch.manual_seed(2024)

class NNRegressor(torch.nn.Module):
    def __init__(self, dims, activation=torch.nn.Tanh(), last_act=None):  # Tanh()  Sigmoid()  ReLU()
        super(NNRegressor, self).__init__()
        self.network = torch.nn.ModuleList([])
        self.dropout = nn.Dropout(0.2)
        for i in range(len(dims)-1):
            self.network.append(torch.nn.Linear(dims[i], dims[i+1]))
            if i != len(dims)-2:
                self.network.append(activation)
#                 self.network.append(self.dropout)
        if not last_act is None:
            self.network.append(last_act)

    def forward(self, x):
        for f in self.network:
            x = f(x)
        return x
```
